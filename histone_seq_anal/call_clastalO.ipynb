```python
import subprocess
import os

current_dir = os.getcwd()
display(current_dir)

def call_clustalo_omega(input_fasta, output_fasta):

    command = ["clustalo", "-i", input_fasta, "-o", output_fasta, "--outfmt", "fasta", '--force']
    try:
        res = subprocess.run(command, check=True, cwd= current_dir)
    except Exception as E:
        display(E)
        print(f"command = {command}")
    else:
        print(f"return code = {res.returncode}")

call_clustalo_omega("data\\H2A\\H2A.fasta", "data\\H2A\\H2A_result.fasta")
```
