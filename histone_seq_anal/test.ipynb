```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots()

# 示例数据
x = range(0, 10)
y = [i**2 for i in x]
ax.plot(x, y)

# 设置 x 轴和 y 轴的刻度间距
# ax.xaxis.set_major_locator(MultipleLocator(2))  # x 轴每隔 2 个单位一个刻度
ax.yaxis.set_major_locator(MultipleLocator(5))  # y 轴每隔 10 个单位一个刻度

plt.show()
```
