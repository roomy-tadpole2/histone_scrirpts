```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Plotting the data
plt.plot(x, y)

# Set custom y-tick labels
y_ticks = [10, 20, 25, 30]
y_tick_labels = ['Low', 'Medium', 'High', 'Very High']
plt.yticks(ticks=y_ticks, labels=y_tick_labels)

# Set colors for the y-tick labels
ax = plt.gca()  # Get the current axis
for tick_label in ax.get_yticklabels():
    tick_label.set_color('red')  # Set color for each y-tick label

# Display the plot
plt.show()
```
